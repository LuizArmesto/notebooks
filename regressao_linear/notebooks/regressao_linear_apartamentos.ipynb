# Regressão Linear

Vamos começar com uma aplicação simples de regressão linear


## Encontrar experimentalmente a constante elástica de uma mola

Temos as medições do quanto a mola estica sob a força exercida por 13 diferentes pequenas massas conhecidas.
Sabemos que, estando em equilíbrio (sem oscilações), a força exercida pela mola é igual ao peso das massa.

Sabendo a fórmula da força peso em função da massa

$$P = m g$$

e da força elástica

$$ F = k \Delta l $$
$$ F = k (l - l _0) $$
$$ F = k l - k l _0) $$

$ k $ é a constante elástica que queremos encontrar, $ k l_0 $ é uma constante (pois $ k $ é constante e $ l_0 $ também é constante, pois é o comprimento da mola sem sofrer tração), portanto temos uma função de reta.

$ y = a + b x $, onde

$$ y = F $$
$$ a = - k l_0 $$
$$ b = k $$


```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np


g = 9.7863808
massa =  np.array([0.0261, 0.0264, 0.0269, 0.0331, 0.0323, 0.0331, 0.0385, 0.0384, 0.1127, 0.0460, 0.0588, 0.0527, 0.0127]) 
peso = np.array([m * g for m in massa])

forca = peso
delta_l = np.array([0.216,  0.217,  0.218,  0.227,  0.226,  0.228,  0.236,  0.236,  0.358,  0.248,  0.269,  0.259,  0.193 ])
```

```python
x = delta_l
y = forca
```

```python
plt.scatter(x, y)
plt.ylabel('Forca (dependente)')
plt.xlabel('Variacao do comprimento da mola em funcao da forca (independente)')
plt.show()
```

Vamos calcular os coeficientes da reta usando o método dos mínimos quadrados

$$ a = \bar y - b \bar x $$

$$ b = \frac{\sum_{n=1}^{n} (x_i - \bar x)(y_i - \bar y)}{\sum_{n=1}^{n} (x_i - \bar x)^2} $$

```python
sxy = np.sum((x - x.mean()) * (y -y.mean()))
syy = np.sum((y - y.mean())**2)
sxx = np.sum((x - x.mean())**2)

B = sxy / sxx
A = y.mean() - B * x.mean() 

print 'y = {} + {}*x'.format(A, B)
```

Encontramos a função 

$$ y = -1.03318008746 + 5.97044972632 x $$

Temos $ a = 5.97044972632 $, portanto $ k = 5.97044972632 $


Neste caso o que nos interessa não é a função em sim, mas apenas o coeficiente agular.


Vamos verificar o coeficiente de determinação para avaliar a qualidade do modelo obtido

```python
print 'R2 score: {}'.format(B*sxy/syy)
```

```python
line_x = [i/100.0 for i in range(18, 38)]
line_y = [(-1.03318008746 + 5.97044972632*x_) for x_ in line_x]
plt.plot(line_x, line_y)
plt.scatter(x, y)
plt.show()
```

## Valor do aluguel de apartamentos em São Paulo

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
```

Temos um CSV com algumas dezenas de milhares de apartamentos disponíveis para locação na cidade de São Paulo, com os valores do aluguel, área do apartamento, quantidade de quartos, de suites e de vagas na garagem e o bairro em que o apartamento se localiza. Os dados foram obtidos por raspagem de um site de classificados.

```python
apartamentos = pd.read_csv('../apartamentos_alugar_sao_paulo.csv')
```

## Valor do aluguel em relação à quantidade de quartos

Uma primeira abordagem é tentar uma relação entre o valor do aluguel pedido e a quantidade de quartos do apartamento

```python
x = apartamentos['quartos']
y = apartamentos['valor']
```

```python
plt.scatter(x, y)
plt.ylabel('Valor (dependente)')
plt.xlabel('Numero de quartos (independente)')
plt.show()
```

```python
print x.shape, y.shape
```

```python
regr = linear_model.LinearRegression()
regr.fit(x.reshape(28879,1), y.reshape(28879,1))
```

# Regressão Linear com mais variáveis

## Valor do aluguel na Vila Madalena em relação a area, quantidade de quartos, de suites e de vagas

```python
cols = ['area', 'quartos', 'suites', 'vagas']
X = apartamentos_bairro[cols]
y = apartamentos_bairro['valor']
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)

# print intercept and coefficients
coef = zip(regr.coef_, cols)
print 'model is "valor = {:.2f} + {:.2f}.{} + {:.2f}.{} + {:.2f}.{}"'.format(regr.intercept_, coef[0][0], coef[0][1], coef[1][0], coef[1][1], coef[2][0], coef[2][1])
print 'r2 = ', r2_score(y, regr.predict(X))
```

O modelo está ainda melhor usando mais variáveis. Vamos tentar prever o valor de dois apartamentos na região.

 - um apartamento de **72 $ m^2 $, 1 quarto, sendo uma suite, e uma vaga na garagem**.
 - um apartamento de **150 $ m^2 $, 3 quartos, sendo 2 suites, com 2 vagas na garagem**.

```python
test = np.array([[72, 1, 1, 1], [150, 3, 2, 2]])
print 'A: {}'.format(regr.predict(test.reshape(2, 4)))
```

Para o primeiro obtivemos a previsão de um aluguel de **R\$ 3.455,96** e para o segundo de **R\$ 5.421,50**

```python
cols = ['area', 'suites']
X = apartamentos_bairro[cols]
y = apartamentos_bairro['valor']

regr = linear_model.LinearRegression()
regr.fit(X, y)

# print intercept and coefficients
coef = zip(regr.coef_, cols)
print 'model is "valor = {:.2f} + {:.2f}.{} + {:.2f}.{}"'.format(regr.intercept_, coef[0][0], coef[0][1], coef[1][0], coef[1][1])
print 'r2 = ', r2_score(y, regr.predict(X))
```

```python
from mpl_toolkits.mplot3d import Axes3D


## Create the 3d plot -- skip reading this
# TV/Radio grid for 3d plot
xx1, xx2 = np.meshgrid(np.linspace(X.area.min(), X.area.max(), 100), 
                       np.linspace(X.suites.min(), X.suites.max(), 100))

# plot the hyperplane by evaluating the parameters on the grid
Z = regr.intercept_ + (regr.coef_[0] * xx1) + (regr.coef_[1] * xx2)

# create matplotlib 3d axes
fig = plt.figure(figsize=(12, 8))
ax = Axes3D(fig, azim=-115, elev=15)

# plot hyperplane
surf = ax.plot_surface(xx1, xx2, Z, cmap=plt.cm.RdBu_r, alpha=0.6, linewidth=0)

# plot data points - points over the HP are white, points below are black
resid = y - regr.predict(X)
ax.scatter(X[resid >= 0].area, X[resid >= 0].suites, y[resid >= 0], color='black', alpha=1.0, facecolor='white')
ax.scatter(X[resid < 0].area, X[resid < 0].suites, y[resid < 0], color='black', alpha=1.0)

# set axis labels
ax.set_xlabel('Area')
ax.set_ylabel('Suites')
ax.set_zlabel('Valor')

```
